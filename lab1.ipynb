```python
def zad3(x1, x2):
    x = np.arange(x1, x2, 0.01)
    plt.figure()
    plt.plot(x, x*x + 5)
    plt.title('Dla x od {0} do {1}'.format(x1, x2))
    plt.xlabel("O x")
    plt.ylabel("O y")
    plt.grid()
    plt.legend(["f(x) = x^2 + 5"], prop={'size':10})
    plt.show()

zad3(-1, 1)
zad3(-6, 6)
zad3(0, 5)
```
